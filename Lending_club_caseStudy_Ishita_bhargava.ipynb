#### data understanding criteria for DTI
> If a borrower makes $5,000 per month and has $1,000 in debt payments (excluding mortgage and the new loan):
DTI = $1,000 Ã· $5,000 = 0.2 or 20%.
This means 20% of the borrowerâ€™s monthly income is already being used to pay debts, leaving 80% available for other expenses or new debt.

> ASSUMPTIO- 10% to 15% DTI is good as customer can take more loans and also can payoff. more % DTI means person is high risk as customer can default.
